```python
with open("input13.txt", "r") as f:
    file = f.read().splitlines()

print(file)
```

```python
timestamp = int(file[0])
buses = [int(i) for i in file[1].split(",") if i != "x"]

earliestTimes = []
for bus in buses:
    mult = 2
    earliest = bus
    while earliest < timestamp:
        earliest = bus*mult
        mult += 1
    earliestTimes.append(earliest)

mostEarliest = min(earliestTimes)
busId = buses[earliestTimes.index(mostEarliest)]
waitingTime = mostEarliest-timestamp
print(busId*waitingTime)
```

```python
# vielfaches von zahl x
# vielfaches von zahl x + an == vielfaches von zahl y
# vielfaches von zahl y + bn == vielfaches von zahl z
buses = file[1].split(",")
buses_dist = {}
buses_mult = {}
dist = -1
init_bus = int(buses[0])
for bus in buses:
    dist += 1
    if bus != "x":
        bus = int(bus)
        buses_dist[bus] = dist
        buses_mult[bus] = 1
    
print(buses_dist)


while not all([True for bus_ind, bus in enumerate(buses_mult.keys()) if bus*buses_mult[bus] == list(buses_mult.keys())[bus_ind-1]*buses_mult[bus]+buses_dist[bus]]):
    for bus, bus_mult in buses_mult.items():
        if bus == init_bus:
            continue
        while bus*buses_mult[bus] != init_bus*buses_mult[init_bus]+buses_dist[bus] and bus*buses_mult[bus] < init_bus*buses_mult[init_bus]:
            buses_mult[bus] += 1

print(buses_mult[init_bus]*init_bus)
```

```python
buses = file[1].split(",")
buses_dist = {}

dist = -1
init_bus = int(buses[0])
for bus in buses:
    dist += 1
    if bus != "x":
        bus = int(bus)
        buses_dist[bus] = dist
    
print(buses_dist)

max_bus = max(buses_dist.keys())

found = False
init_x = 23
x = init_x
while not found:
    print(x)
    if x-buses_dist[max_bus]%init_bus == 0:
        worked = True
        for bus, dist in buses_dist.items():
            if x+dist%bus != 0:
                worked = False
                break
        if worked:
            break
    x *= 2
print(x)
```
